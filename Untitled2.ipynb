```python
# vasavi sandhya @CSU & vinitha @CSU
import pandas as pd
#Load the dataset
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex',
           'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
data = pd.read_csv("adult.csv", names=columns, na_values='?', skipinitialspace=True)

random_sample = data.sample(n=1000)
stratified_sample = data.groupby('education').apply(lambda x: x.sample(frac=0.1))
#calculate key statistics for the population ans samples
statistics_df = pd.DataFrame( {
    'Population': data['hours-per-week'].describe(),
    'Simple Random': random_sample['hours-per-week'].describe(),
    'stratified': stratified_sample['hours-per-week'].describe()
})
print(statistics_df)
```

```python
# vasavi sandhya @CSU & vinitha @CSU
statistics_df.loc[['mean','50%','std']].plot(kind='bar', figsize=(10,6))
plt.title('Comparison of Key Statistics: Population vs.Samples')
plt.ylabel('value')
plt.xticks(rotation=0)
plt.legend(['Population','Simple Random','Stratified'])
plt.show()
```

```python
# vasavi sandhya @CSU & vinitha @CSU
import seaborn as sns
import matplotlib.pyplot as plt

# Resetting index for stratified sample to ensure consistency
stratified_sample = stratified_sample.reset_index(drop=True)


# Creating a combined DataFrame to hold data from both samples
combined_data = pd.DataFrame({
    'Simple Random Sample': random_sample['hours-per-week'],
    'Stratified Sample' : stratified_sample['hours-per-week']
})
#visualize the distribution of 'hours_per_week' across  the two samples
plt.figure(figsize=(10,6))
#sns.boxplot(data=[random_sample['hours-per-week'], stratified_sample['hours-per-week']])
sns.boxplot(data=combined_data)
plt.xticks([0,1], ['Simple Random Sample', 'Stratified Sample'])
plt.ylabel('Hours Per Week')
plt.show()
```

```python
# vasavi sandhya @CSU & vinitha @CSU
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

iris = load_iris()
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns= iris['feature_names'] +['target'])

#convert numeric target to species name for clarity
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
df.drop('target', axis=1, inplace=True)
```
